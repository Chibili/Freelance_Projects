# Linearity Assumptions

```python
# Load Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy as sp
import seaborn as sns
import tensorflow as tf
import statsmodels.api as sm
import statsmodels.tsa.api as smt
import warnings

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor

from pylab import rcParams
import matplotlib.pyplot as plt
from matplotlib import rc
from sklearn.model_selection import train_test_split
import joblib

%matplotlib inline
%config InlineBackend.figure_format='retina'
sns.set(style='whitegrid', palette='muted', font_scale=1.5)
rcParams['figure.figsize'] = 16, 10
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)
```

### Load and Split the Data

```python
df = pd.read_csv('CO2_Cleaned.csv')

target_col = "CO2"

X = df.loc[:, df.columns != target_col]
y = df.loc[:, target_col]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RANDOM_SEED)
```

```python
from sklearn.preprocessing import StandardScaler
from tensorflow.keras import models, layers

# Set random seed
from numpy.random import seed
seed(42)
import tensorflow
tensorflow.random.set_seed(42)

scaler = StandardScaler()
scaler.fit(X_train)
x_train = scaler.transform(X_train)
x_test = scaler.transform(X_test)
```

#### Number of Rows and Columns

```python
print ("Total number of rows in dataset = {}".format(df.shape[0]))
print ("Total number of columns in dataset = {}".format(df.shape[1]))
```

### Linear Regression Using Statsmodels

```python
X_with_constant = sm.add_constant(X_train)
model = sm.OLS(y_train, X_with_constant)
```

```python
results = model.fit()
results.params
```

```python
print(results.summary())
```

```python
X_test = sm.add_constant(X_test)
y_pred = results.predict(X_test)
residual = y_test - y_pred
```

### Multicolinearity Check

```python
vif = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
pd.DataFrame({'vif': vif[0:]}, index=X_train.columns).T
```

### Normailty of Residual Verification

```python
sns.distplot(residual)
```

```python
# Q-Q Plot
import scipy as sp

fig, ax = plt.subplots(figsize=(6,4.5))
_, (__, ___, r) = sp.stats.probplot(residual, plot=ax, fit=True)
```

```python
# Verification that the distribution is normal because the residual value is less than 0
np.mean(residual)
```

###  Homoscedasticity

```python
fig, ax = plt.subplots(figsize=(7,6.5))
_ = ax.scatter(y_pred, residual)
```

### No Autocorrelation of Residuals

```python
auto_c = smt.graphics.plot_acf(residual, lags=40 , alpha=0.05)
auto_c.show()
```
